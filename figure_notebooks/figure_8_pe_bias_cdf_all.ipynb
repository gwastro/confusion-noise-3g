```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import hist
```

```python
pe_bias_CE_median = np.loadtxt("./pe_bias_CE_median.txt", unpack=True, skiprows=1)
pe_bias_CE_upper = np.loadtxt("./pe_bias_CE_upper.txt", unpack=True, skiprows=1)
pe_bias_ETD_median = np.loadtxt("./pe_bias_ETD_median.txt", unpack=True, skiprows=1)
pe_bias_ETD_upper = np.loadtxt("./pe_bias_ETD_upper.txt", unpack=True, skiprows=1)
```

```python
f = plt.figure(figsize=(6, 8), dpi=150)
ax1 = f.add_subplot(211)
ax2 = f.add_subplot(212)

bins_method = 'freedman'
xmax1 = 0.003
ax1.hlines(y=1, xmin=0, xmax=xmax1, linestyles="solid", colors="gray")
hist(pe_bias_CE_median[0], ax=ax1, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='solid', color='#2CA02C', label=r"CE BNS (median $R_{0}$)")
hist(pe_bias_CE_upper[0], ax=ax1, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='dashed', color='#2CA02C', label=r"CE BNS (upper $R_{0}$)")
hist(pe_bias_ETD_median[0], ax=ax1, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='solid', color='#1F77B4', label=r"E1 BNS (median $R_{0}$)")
hist(pe_bias_ETD_upper[0], ax=ax1, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='dashed', color='#1F77B4', label=r"E1 BNS (upper $R_{0}$)")
ax1.set_xlim(0, xmax1)
ax1.set_ylim(0, 1.05)
ax1.set_xlabel(r"$\left|\mathcal{M}_{0}-\mathcal{M}_{m}\right| / \mathcal{M}_{0}$", fontsize=11)
ax1.set_ylabel("CDF", fontsize=11)
ax1.tick_params(axis = 'both', which = 'major', labelsize = 8)
ax1.legend(fontsize=8.1, loc='lower right')

xmax2 = 0.03
ax2.hlines(y=1, xmin=0, xmax=xmax2, linestyles="solid", colors="gray")
hist(pe_bias_CE_median[1], ax=ax2, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='solid', color='#2CA02C', label=r"CE BNS (median $R_{0}$)")
hist(pe_bias_CE_upper[1], ax=ax2, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='dashed', color='#2CA02C', label=r"CE BNS (upper $R_{0}$)")
hist(pe_bias_ETD_median[1], ax=ax2, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='solid', color='#1F77B4', label=r"E1 BNS (median $R_{0}$)")
hist(pe_bias_ETD_upper[1], ax=ax2, density=True, bins=bins_method, histtype='step', cumulative=True, linestyle='dashed', color='#1F77B4', label=r"E1 BNS (upper $R_{0}$)")
ax2.set_xlim(0, xmax2)
ax2.set_ylim(0, 1.05)
ax2.set_xlabel(r"$\left|t_{0}-t_{m}\right|$ (s)", fontsize=11)
ax2.set_ylabel("CDF", fontsize=11)
ax2.tick_params(axis = 'both', which = 'major', labelsize = 8)
ax2.legend(fontsize=8.1, loc='lower right')

f.savefig('pe_bias_cdf_all.pdf', dpi=500)
f.show()
```
